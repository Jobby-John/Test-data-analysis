Data analysis

```python
#first we have to come up with an age column in our dataset
#first we have to date of birth column to date format
#Encountered an error so we had to change the year to a correct format
data['date_of_birth'] = pd.to_datetime(data['date_of_birth'])
```

```python
data.head()
```

```python
# Add the age column in the dataset
data['Age']= data["date_of_birth"].apply(lambda x : (pd.datetime.now().year - x.year))
```

```python
data.head(31)
```

DATA ANALYSIS

```python
#To show distribution of male and female in the dataset
genderD = data.groupby('gender').size().rename_axis('gender_Distribution').reset_index(name='Count')
genderD.head()
```

```python
# A piechart to show the ratio of gender
data.groupby('gender').size().plot(kind='pie', autopct='%.2f')
plt.title("Distribution by Gender", size=15)
plt.show();
```

```python
# A horizontal histogram to show the distribution of gender
data.groupby('gender').size().plot(kind='barh')
plt.title("Distribution by Gender", size=15)
plt.show();
```

```python
#To show distribution of Activity in the dataset
Activity_D = data.groupby('activity_name').size().rename_axis('Activity').reset_index(name='Count')
Activity_D.head()
```

```python
# A pie chart to show activity distribution
data.groupby('activity_name').size().plot(kind='pie', autopct='%.2f')
plt.title("Distribution by Activity", size=15)
plt.show();
```

```python
#A horizontal histogram to show the distribution of Activities
data.groupby('activity_name').size().plot(kind='barh')
plt.title("Distribution by Activity", size=15)
plt.show();
```
